## The Trigonometric functions

Use the mnemonic: `SOH CAH TOA`

$\sin(\theta) = \text(opposite) / \text(hypotenuse)$

$\cos(\theta) = \text(adjacent) / \text(hypotenuse)$

$\tan(\theta) = \text(opposite) / \text(adjacent)$


The result of each function is a ratio. The *tangent* of an angle is the vertical distance covered divided by the horizontal. In comparison, *sine* and *cosine* is the vertical and horizontal distance covered relative to the overall distance (hypotenuse).

***Sine* and *cosine* are percentages.** They vary from +100% to 0 to -100%.

Sine is height, as a percentage of the hypotenuse. Cosine is the width, as a percentage of the hypotenuse.

As the degree reaches 90°, sine reaches 100%, and cosine reaches 0%.
As the degree reaches 0°, sine reaches 0%, and cosine reaches 100%.

Given an angle, the coordinates of vectors along that angle will have a constant ratio: the *tangent* of the angle.

```python
import math

def get_trig_percentages(degree):
    print(f"Degrees: {degree}°")
    degree = math.radians(degree)
    sine = math.sin(degree) * 100
    cosine = math.cos(degree) * 100
    print(f"sine: {sine:.1f}% of height")
    print(f"cosine: {cosine:.1f}% of width")
    print()

get_trig_percentages(0)
get_trig_percentages(1)
get_trig_percentages(30)
get_trig_percentages(90)
get_trig_percentages(130)
```

### With known angle

Given an angle of 37°, with a hypotenuse of 5:

For every 5 units traveled at 37°, 3 vertical units are traveled: $\sin(37°) = 3/5 = 0.6$

For every 5 units traveled at 37°, 4 horizontal units are traveled: $\cos(37°) = 4/5 = 0.8$

### Polar to cartesian

Given a known angle $\theta$ and distance $r$ (hypotenuse) the cartesian components are:

$x = r * \cos(\theta)$

$y = r * \sin(\theta)$

## Radians

1 radian $\approx$ 57.296°

$\pi$ radians = 180°

$2\pi$ radians = 360°

```python
from math import pi

def radians_to_degrees(rad):
    return rad / pi * 180

def degrees_to_radians(deg):
    return deg / 180 * pi
```

```python
from math import pi, sin, cos

# polar_vector = (length, degree)
def to_cartesian(polar_vector):
    length = polar_vector[0]
    angle = polar_vector[1]
    return (length * cos(angle), length * sin(angle))


# angle to radians
# get num of radians in degree
radians_in_degree = pi / 180
angle = 38 * radians_in_degree


to_cartesian((5, angle))
```
