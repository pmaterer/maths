## Addition

$$\vec{A} + \vec{B} = \vec{C}$$

$$a = (1, 5)$$ $$b = (-2, 7)$$

$$c = (1 + -2), (5 + 7)$$

$$c = (-1, 12)$$

```python
from vector import Vector

v1 = Vector(1, 5)
v2 = Vector(-2, 7)
v3 = v1 + v2
print(v3)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.red)
a2 = Arrow(v2, color=colors.red)
a3 = Arrow(v3, color=colors.blue)

# tip -> tail
a4 = Arrow(tip=v2, tail=v1, color=colors.purple)
a5 = Arrow(tip=v1, tail=v2, color=colors.purple)

draw(a1, a2, a3, a4, a5, width=4)
```

## Scalar Multiplication

$$\vec{a} = (1.2, -3.1)$$

$$scalar = 6.5$$

$$6.5 * \vec{a} = 6.5 * (1.2, -3.1) = (6.5 * 1.2, 6.5 * -3.1) = (7.8, -20.15)$$

```python
from vector import Vector

v1 = Vector(1.2, -3.1)
v2 = v1 * 6.5
print(v2)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.black)
a2 = Arrow(v2)

draw(a2, a1)
```

## Length

Use the Pythagorean theorem: $$a^2 + b^2 = c^2$$

$$\vec{A} = (-3, 7)$$

$$\sqrt{3^2 + 7^2} = \sqrt{9 + 49} = \sqrt{58} = 7.6157 ...$$

```python
from vector import Vector
v1 = Vector(-3, 7)
print(v1.length())
```

## Subtraction

$$\vec{A} - \vec{B} = \vec{C}$$

$$a = (-1, 3)$$ $$b = (2, 2)$$

$$c = (-1 - 2), (3 - 2)$$

$$c = (-3, 1)$$

```python
from vector import Vector

v1 = Vector(-1, 3)
v2 = Vector(2, 2)
v3 = v1 - v2
print(v3)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.red)
a2 = Arrow(v2, color=colors.red)
a3 = Arrow(v3, color=colors.blue)

a4 = Arrow(tip=v3, tail=v2, color=colors.blue)

draw(a1, a2, a3, a4, width=4)
```
