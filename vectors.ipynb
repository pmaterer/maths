## Addition

$$\vec{A} + \vec{B} = \vec{C}$$

$$a = (1, 5)$$ $$b = (-2, 7)$$

$$c = (1 + -2), (5 + 7)$$

$$c = (-1, 12)$$

```python
from vector import Vector

v1 = Vector(1, 5)
v2 = Vector(-2, 7)
v3 = v1 + v2
print(v3)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.red)
a2 = Arrow(v2, color=colors.red)
a3 = Arrow(v3, color=colors.blue)

# tip -> tail
a4 = Arrow(tip=v2, tail=v1, color=colors.purple)
a5 = Arrow(tip=v1, tail=v2, color=colors.purple)

draw(a1, a2, a3, a4, a5, width=4)
```

## Scalar Multiplication

$$\vec{a} = (1.2, -3.1)$$

$$scalar = 6.5$$

$$6.5 * \vec{a} = 6.5 * (1.2, -3.1) = (6.5 * 1.2, 6.5 * -3.1) = (7.8, -20.15)$$

```python
from vector import Vector

v1 = Vector(1.2, -3.1)
v2 = v1 * 6.5
print(v2)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.black)
a2 = Arrow(v2)

draw(a2, a1)
```

## Length

Use the Pythagorean theorem: $$a^2 + b^2 = c^2$$

$$\vec{A} = (-3, 7)$$

$$\sqrt{3^2 + 7^2} = \sqrt{9 + 49} = \sqrt{58} = 7.6157 ...$$

```python
from vector import Vector
v1 = Vector(-3, 7)
print(v1.length())
```

## Subtraction (displacement)

$$\vec{A} - \vec{B} = \vec{C}$$

$$a = (-1, 3)$$ $$b = (2, 2)$$

$$c = (-1 - 2), (3 - 2)$$

$$c = (-3, 1)$$

```python
from vector import Vector

v1 = Vector(-1, 3)
v2 = Vector(2, 2)
v3 = v1 - v2
print(v3)
```

```python
import colors
from shapes import Arrow
from draw import draw

a1 = Arrow(v1, color=colors.red)
a2 = Arrow(v2, color=colors.red)
a3 = Arrow(v3, color=colors.blue)

a4 = Arrow(tip=v3, tail=v2, color=colors.blue)

draw(a1, a2, a3, a4, width=6)
```

## Negate

Give a vector v, the negative vector -v is the same as the scalar multiple of -1 * v.


```python
v1 = Vector(1, 1)
v2 = v1.negate()
```

```python
from shapes import Arrow
from draw import draw
a1 = Arrow(v1)
a2 = Arrow(v2)
draw(a1, a2)
```

## Translate

Use a translation vector on a list of vectors (polygon).


```python
import colors
from vector import Vector, distance
from shapes import Polygon
from draw import draw

polygon_vertices = [
    Vector(-2, 5),
    Vector(5, 2),
    Vector(2, -3),
    Vector(-3, -2),
]

polygon = Polygon(*polygon_vertices)
polygon_translated = polygon.translate(Vector(-1,1))
polygon_translated.color = colors.red

draw(polygon, polygon_translated)

print(f"perimeter: {polygon.perimeter()}")
```

## Rotations

In polar coordinates, adding to the angle rotates a vector further counterclockwise. Subtracting from it rotates it clockwise.

```python
import colors
from vector import Vector, rotate
from shapes import Polygon, Arrow
from draw import draw
from conversions import to_cartesian, to_polar
from math import radians, pi

# rotation_angle = radians(90)
rotation_angle = pi/4

polygon_vertices = [
    Vector(-2, 5),
    Vector(5, 2),
    Vector(2, -3),
    Vector(-3, -2),
]
rotated_polygon_vertices = rotate(pi/4, polygon_vertices)

print(polygon_vertices)
print(rotated_polygon_vertices)

draw(Polygon(*polygon_vertices), Polygon(*rotated_polygon_vertices, color=colors.red))
```

## Dot Product

The dot product of two vectors can be seen as the project of one over the other. The result relates to the cosine of the angle. ls

**Method One**

$a · b = |a| * |b| * \cos(\theta)$

Where $|x|$ is the magnitude (length) of a vector.

**Method Two**

$a · b = (a_{x} * b_{x}) + (a_{y} * b_{y})$


### Examples

**Method One**

$a · b = 10 * 13 * \cos(59.5°)$

$a · b = 10 * 13 * 0.5075...$

$a · b = 65.98... = 66$

**Method Two**

$a · b = (-6 * 5) + (8 * 12)$

$a · b = -30 + 96$

$a · b = 66$


```python
import colors
from vector import Vector
from shapes import Points, Arrow
from draw import draw

points = Points(
    Vector(0, 4, 2),
    Vector(-2, 1),
    Vector(1, -3, -2)
)

draw(points, width=5)

v1 = Vector(-2, 1)
a1 = Arrow(v1)
draw(a1)
```
